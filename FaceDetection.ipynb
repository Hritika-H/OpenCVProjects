#Face detection using OpenCV

```python
import cv2 as cv
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

import image and haarcascade_frontalface_alt.xml

```python
original_image=cv.imread('/content/familypic.jpeg')
plt.imshow(cv.cvtColor(original_image, cv.COLOR_BGR2RGB))
```

```python
grayscale_image=cv.cvtColor(original_image,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv.cvtColor(grayscale_image, cv.COLOR_BGR2RGB))
original_image.shape
```

```python
face_cascade=cv.CascadeClassifier('/content/haarcascade_frontalface_alt.xml')
```

```python
detected_faces = face_cascade.detectMultiScale(grayscale_image)
```

```python
for(column,row,width,height) in detected_faces:
  cv.rectangle(original_image,(column, row),(column + width, row + height),(0, 255, 0),2)
```

```python
plt.imshow(cv.cvtColor(original_image, cv.COLOR_BGR2RGB))
```
